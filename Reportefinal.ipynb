## I. Universidad Nacional de Piura

### 1. Leyendo el primer repositorio de la Universidad Nacional de Piura

```python
#  1. Requerimiento 

website = 'https://repositorio.unp.edu.pe/handle/20.500.12676/3728?show=full'
print('Haciendo una solicitud a:\n')
print(website)

unp = requests.get(website)
unp #Indica el status. 200 significa 'Respuesta satisfactoria'
```

```python
#  2. Scrapear con BS

    # a. Almacenar el texto en una 'sopa'
    
soup = bs(unp.text,'lxml') #'lxml' es el parse, el contenido lo divide en pedazos y los almacena en una "sopa-"
  
    # c. Imprimir texto
print(soup.prettify())
```

### 3. Extraemos la información de la primera tesis 

```python
caja = soup.find('div', class_="ds-table-responsive") ### Establecemos los parámetros de la información que vamos a utilizar

### De la "caja" extraemos la información necesaria guíandonos del texto del cuadro y extrayendo el texto con la información
titulo = caja.find("td", text = "dc.title").find_next_sibling("td").text 

asesor = caja.find("td", text = "dc.contributor.advisor").find_next_sibling("td").text

autor = caja.find("td", text = "dc.contributor.author").find_next_sibling("td").text

resumen = caja.find("td", text = "dc.description.abstract").find_next_sibling("td").text

uni = caja.find("td", text = "dc.source").find_next_sibling("td").text

año = caja.find("td", text = "dc.date.issued").find_next_sibling("td").text

grado = caja.find("td", text = "thesis.degree.name").find_next_sibling("td").text
```

```python
import pandas as pd
# Generando el dataframe

df = pd.DataFrame({'Institución': [uni],
                   'Título': [titulo],
                   'Autor': [autor],
                   'Grado': [grado],
                   'Asesor': [asesor],
                   'Resumen': [resumen],
                   'Año': [año]})
df
```

### 4. Automatización de la descarga en csv

```python
df.to_csv('Reportefinal', encoding = 'utf-8')
#Gracias a la herramienta de pandas hemos pasado el dataframe a un archivo con la extensión csv,
#para su posterior análisis en excel.
```

## II. Universidad Miguel Hernandez de Elche

### 1. Leyendo el repositorio de Tesis de Fin de Grado de Estadística Empresarial

```python
#  1. Requerimiento 

website_umh = 'http://dspace.umh.es/handle/11000/1190'
print('Haciendo una solicitud a:\n')
print(website_umh)

listado_umh = requests.get(website_umh)
listado_umh #Indica el status. 200 significa 'Respuesta satisfactoria'
```

```python
#  2. Scrapear con BS

    # a. Almacenar el texto en una 'sopa'
    
sopa_umh = bs(listado_umh.text,'lxml') #'lxml' es el parse que pide la funcion

    # b. Imprimir texto
print(sopa_umh.prettify())
```

### 2. Extraemos el url de cada tesis

```python
# Extraer url de cada tesis 

tesis = sopa_umh.find('table').find_all('td', attrs = {'headers':'t3'})

# Obteniendo los urls y generando una lista con ellos

urls_umh=[]

for url in tesis:
    a=website_umh[0:20]+url.a.get('href')
    urls_umh.append(a)
        
urls_umh
```

### 3. Extraemos la información de cada tesis

```python
#  1. Hacer el requerimiento a cada tesis en un bucle for, y extraer con BS la informacion requerida de cada tesis.

#  a. Creando listas vacias para almacenar la informacion.

titles=[]
authors=[]
advisors=[]
abstracts=[]
university=[]
grade=[]
year=[]

#  b. Haciendo requerimientos y extracciones

for url in urls_umh:
    a=requests.get(url)
    if a.status_code == 200:
        print('Solicitud a ', url,' satisfactoria\n')
        soup2=bs(a.text,'lxml')

        # Extraigo los aspectos de la tesis
        titles.append(soup2.find("span", attrs = 'metadataFieldValue').text)
        authors.append(soup2.find("a", attrs = 'author').text)
        advisors.append((soup2.find('table').find_all('span', attrs = 'metadataFieldValue'))[2].text)
        abstracts.append((soup2.find('table').find_all('span', attrs = 'metadataFieldValue'))[6].text)
        university.append((soup2.find('table').find_all('span', attrs = 'metadataFieldValue'))[3].text)
        grade.append((soup2.find('ol', attrs = 'breadcrumb').find_all('li'))[3].text)
        year.append((soup2.find('table').find_all('span', attrs = 'metadataFieldValue'))[5].text[0:4])

print('Información extraída\n')
```

```python
#  2. Generando el dataframe con cada aspecto de todas las tesis

df_total_umh = pd.DataFrame(zip(university,titles,authors,grade,advisors,abstracts,year),
                        columns=['Institución','Título','Autor','Grado','Asesor','Resumen','Año']
                        )
df_total_umh
```

### 4. Automatización de la descarga en csv

```python
df_total_umh.to_csv('Tesis_UMH', encoding = 'utf-8')

print('La tabla ha sido descargada con el nombre Tesis_UMH.')
```

## III. Pontificia Universidad Católica del Perú

### 1. Leyendo el repositorio de Tesis de Fin de Grado de Estadística Empresarial

```python
#  1. Requerimiento 

website_pucp = 'https://tesis.pucp.edu.pe/repositorio/handle/20.500.12404/14'
print('Haciendo una solicitud a:\n')
print(website_pucp)

listado_pucp = requests.get(website_pucp)
listado_pucp #Indica el status. 200 significa 'Respuesta satisfactoria'
```

```python
#  2. Scrapear con BS

    # a. Almacenar el texto en una 'sopa'
    
sopa_pucp = bs(listado_pucp.text,'lxml') #'lxml' es el parse que pide la funcion

    # b. Imprimir texto
print(sopa_pucp.prettify())
```

### 2. Extraemos el url de cada tesis

```python
# Extraer url de cada tesis 

tesis = sopa_pucp.find('ul', attrs = 'ds-artifact-list list-unstyled').find_all('h4')

# Obteniendo los urls y generando una lista con ellos

urls_pucp=[]

for url in tesis:
    a=website_pucp[0:25]+url.a.get('href')+'?show=full'
    urls_pucp.append(a)
        
urls_pucp
```

### 3. Extraemos la información de cada tesis

```python
#  1. Hacer el requerimiento a cada tesis en un bucle for, y extraer con BS la informacion requerida de cada tesis.

#  a. Creando listas vacias para almacenar la informacion.

titles=[]
authors=[]
advisors=[]
abstracts=[]
university=[]
grade=[]
years=[]

#  b. Haciendo requerimientos y extracciones

for url in urls_pucp:
    a=requests.get(url)
    if a.status_code == 200:
        print('Solicitud a ', url,' satisfactoria\n')
        soup2=bs(a.text,'lxml')

        # Extraigo los aspectos de la tesis
        title=soup2.find("h2").text
        titles.append(title)
        
        asesor1 = soup2.find("table").find_all('td', attrs = 'word-break')[0].text
        if len(soup2.find("table").find_all('td', attrs = 'word-break')[4].text) != 4:
            if soup2.find("table").find_all('td', attrs ='label-cell')[1].text == 'dc.contributor.advisor':
                asesor2=soup2.find("table").find_all('td', attrs = 'word-break')[1].text
                advisors.append((asesor1+" y "+asesor2))
                autor=soup2.find("table").find_all('td', attrs = 'word-break')[2].text
            elif soup2.find("table").find_all('td', attrs ='label-cell')[1].text == 'dc.contributor.author' and soup2.find("table").find_all('td', attrs ='label-cell')[2].text == 'dc.contributor.author':
                autor=soup2.find("table").find_all('td', attrs = 'word-break')[1].text + ' y '+ soup2.find("table").find_all('td', attrs = 'word-break')[2].text
            
            year=soup2.find("table").find_all('td', attrs = 'word-break')[5].text
            abstracts.append(soup2.find("table").find_all('td', attrs = 'word-break')[8].text)
            university.append(soup2.find("table").find_all('td', attrs = 'word-break')[10].text)
       
        else:
            autor=soup2.find("table").find_all('td', attrs = 'word-break')[1].text
            year=soup2.find("table").find_all('td', attrs = 'word-break')[4].text
            abstracts.append(soup2.find("table").find_all('td', attrs = 'word-break')[7].text)
            university.append(soup2.find("table").find_all('td', attrs = 'word-break')[9].text)

        advisors.append(asesor1)
        authors.append(autor)
        years.append(year)
        grade.append(soup2.find('ul', attrs = 'breadcrumb hidden-xs').find_all('li')[3].text)
        
print('Información extraída\n')
```

```python
#  2. Generando el dataframe con cada aspecto de todas las tesis

df_total_pucp = pd.DataFrame(zip(university,titles,authors,grade,advisors,abstracts,years),
                        columns=['Institución','Título','Autor','Grado','Asesor','Resumen','Año']
                        )
df_total_pucp
```

### 4. Automatización de la descarga en csv

```python
df_total_pucp.to_csv('Tesis_PUCP', encoding = 'utf-8')

print('La tabla ha sido descargada con el nombre Tesis_PUCP.')
```
